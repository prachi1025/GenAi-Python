# Data Science 4th Class

```python
import pandas as pd
df = pd.DataFrame({'Name': ['prachi', 'chaitnya', 'gunjan', 'ahnis', 'vani', 'muskan', 'mitanshi'], 'Age': [22, 22, 22, 23, 50, 21, 23], 'Class': ['Btech', 'Bca', 'Mtech', 'BCA', 'BCA', 'BCA', 'Mtech']})
df
```

## insert fuction for DataFrames
<p>insert function is used to insert a new column with values to a particular location in the DataFrame.</p>

```python
df.insert(loc = 2, column='marks', value=[45,43,46,44,39,30,20])
df
```

## sort_values Function for DataFrame
<p>sort_values is used to sort the records of a DataFrame according to selected column</p>

```python
df.sort_values(by='marks')
```

```python
df.sort_values(by='marks', ascending=False)
```

```python
df.sort_values(by='Age', ascending=False, ignore_index=True)
```

## info function for DataFrames
<p>This method prints information about a DataFrame including the index dtype and columns, non-null values and memory usage.</p>

```python
df.info()
```

## columns property for DataFrames
<p>This property is used to retreive all the column names of the DataFrame according to their index</p>

```python
df.columns
```

## iterrows function for DataFrames
<p>The iterrows() method in pandas is used to iterate through the rows of a DataFrame, yielding the index and the row data as a Series object for each iteration</p>

```python
for i in df.iterrows():
    print(i)
```

```python
for i, row in df.iterrows():
    print(f"At index {i}, {row['Name']} scored {row['marks']} in {row['Class']}")
```

## describe function for DataFrames
<p>The pandas.DataFrame.describe() method generates descriptive statistics that summarize the central tendency, dispersion, and shape of a dataset's distribution. This function is a quick way to get a high-level overview of your data and understand its key characteristics.</p>

```python
df.describe()
```

```python
df.describe(include='all')
```

<p>count - Number of non-null entries</p>
<p>mean	- Average value</p>
<p>std	- Standard deviation</p>
<p>min	- Minimum value</p>
<p>25%	- 25th percentile (Q1)</p>
<p>50%	- Median (50th percentile)</p>
<p>75%	- 75th percentile (Q3)</p>
<p>max	- Maximum value</p>
<p>unique - Number of unique values</p>
<p>top	- Most frequent value</p>
<p>freq	- Frequency of the most frequent value</p>
