```python
df['Age']
```

```python
df[['Name']]
```

```python
df[1:]
```

## Accessing elements using .loc[] and .iloc[]

```python
df.loc[:3,:] # loc[row index, column index]
```

```python
df.iloc[:3,:2] # iloc[row index, column index]
```

```python
df.loc[:, 'Name'] # loc[row index, column name]
# this returns a Series of the Name column
```

```python
df.loc[:2,['Name','Class']] # loc[row index, [column names]]
# this returns a DataFrame of the Name and Class columns till label 2
```

```python
df.iloc[:,:2] # iloc[row inde, column index]
# this returns a DataFrame of the first two columns
```

```python
df.loc[2,['Name','Class']] # loc[row index, [column names]]
# this returns an Object of the Name and Class columns at index 2
```
