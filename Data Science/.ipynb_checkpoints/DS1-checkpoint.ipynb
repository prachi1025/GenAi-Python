```python
# pandas give us 2 main data structures:
# series - a labeled 1D array.
# DataFrame - a labeled 2D array.
```

```python
# Pandas Series (1D) 

s = pd.Series([10,20,23,56,34,22])
s
```

```python
s1 = pd.Series([12, True, 'Prachi', 100, 5.3])
s1
```

```python
s2 = pd.Series([i for i in range(0,20)])
s2
```

```python
s3 = pd.Series([i**2 for i in range(0,20)], index = [i for i in range(0,20)])
s3
```

```python
s4 = pd.Series(data = [23, True, 'Prachi', 100, 6.5], index = [ 'a', 'b', 'c', 'd', 'e' ])
s4

# Index is NOT just numbering. It is a label.
```

## Accessing Data in Series

### Normal Slicing

```python
# Values of Series can be accessed using normal list slicing []
s2[:3]
```

```python
s1[:2]
```

```python
s4.b

# this works because index is not just numbering but it is label
```

### .loc - label based 
<p> This works on labels and includes end </p>
<p>.loc[:'n'] : selects from label start to label 'n'. It does on indices when no labe;s are assigned, but if labels are assigned then only on labels. </p>

```python
s4.loc[:'b']
```

```python
s3.loc[:5]
```

## .iloc - index position based
<p>This works like normal Python slicing. .iloc excludes the end value. </p>

```python
s4.iloc[:2]
```

```python
s3.iloc[:5]
```

# Conditions 
## step 1 - condition creates Boolean Series

```python
s2 > 5
```

## step 2 - Boolean Masking

```python
s2[s2 > 5]
```

<p>This selects only True rows.</p>

# Multiple Conditions

```python
s2[(s2 > 5) & (s2 < 15)]
```

<p>Must use & instead of and, | instead of or. Because this is vectorized operation on arrays.</p>

# betweem()

```python
s2[s2.between(5,15)]
```

<p>we can use inclusive in between() to define which limit do we want to include in the condition</p>

```python
s2[s2.between(5,15, inclusive='right')]
```

```python
s2[s2.between(5,15, inclusive='left')]
```

```python
s2[s2.between(5,15, inclusive='both')]
```

# isin()
<p>checks if value exists in list.</p>

```python
s2.isin([5])
```

```python
s2[s2.isin([5])]
```

```python
s2[s2.isin([30])]
```
