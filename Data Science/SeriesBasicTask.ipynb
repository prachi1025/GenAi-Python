# Series Basic Task

```python
import pandas as pd
data = pd.Series([100,200,300,400], index=['x','y','z','w'])
print("The element at 'y' index is: ", data.loc['y'])
```

```python
print("The 2nd element in the series is: ", data.iloc[1])
```

```python
print("The elements from 'x' to 'w' indices are: ", "\n",  data.loc['x':'w'])
```

```python
print("The last element in the series is: ", data.iloc[-1])
```

```python
print("The elements from 'y' to 'z' are: ", "\n" ,data.loc['y':'z'])
```

```python
print("The element at 'x' in the Series is: ", data.x)
```
